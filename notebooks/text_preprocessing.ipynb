```python
import pyarabic.araby as araby 
import re
import pandas as pd    
```

```python
_ARABIC_TOKEN_RE = re.compile(r'[\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF0-9]+', re.UNICODE)
def removeStopWords(text, stop_words_path="../src/stop_words_arabic.csv"): 
    """
    Stop-word removal using text re and araby: 
    - loads stop words from file one (cached on function) 
    - strips tashkeel and normalizes ligatures 
    - tokenizes with a compiled Arabic+digits regex
    """
    if not hasattr(removeStopWords, "_stop_words"): 
        s = set() 
        try: 
            with open(stop_words_path, 'r', encoding='utf-8') as f :
                for line in f: 
                    for p in line.strip().split(','): 
                        p = p.strip() 
                        if p: 
                            s.add(p) 
        except FileNotFoundError: 
            s = set()  
        removeStopWords._stop_words = s
    clean = araby.strip_tashkeel(text)  
    clean = araby.normalize_ligature(clean) 
    token = _ARABIC_TOKEN_RE.findall(clean) 
    filtered = [t for t in token if t not in removeStopWords._stop_words] 
    return ' '.join(filtered)

```

```python
## remove Icons and Hashtags 
emoji_pattern = re.compile("["
         u"\U0001F600-\U0001F64F"  # emoticons
         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
         u"\U0001F680-\U0001F6FF"  # transport & map symbols
         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
         u"\U00002702-\U000027B0"
         u"\U000024C2-\U0001F251"
         "]+", flags=re.UNICODE) 
def remove_icons_hashtags(text): 
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+','', text) # remove mentions 
    text = re.sub(r'\#\w+', '', text) # remove hashtags 
    text = re.sub(r'[^\w\s]', '', text) # remove punctuations 
    text = emoji_pattern.sub(r'', text) # Remove emojis 

    return text.strip() 
    
```
